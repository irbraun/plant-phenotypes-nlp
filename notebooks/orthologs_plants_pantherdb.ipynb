```python
import pandas as pd
```

```python
file = "../data/orthology_related_files/pantherdb/PlantGenomeOrthologs"
df = pd.read_table(file, names=["gene","ortholog","type","ancestor","pantherid"])
df["gene_species"] = df["gene"].map(lambda x: x.split("|")[0])
df["ortholog_species"] = df["ortholog"].map(lambda x: x.split("|")[0])
df.head(20)
```

```python
species = ["ARATH","MAIZE","SOLLC","MEDTR","SOYBN","ORYSJ"]
df = df[(df["ortholog_species"].isin(species)) & (df["gene_species"].isin(species))]
df.head(20)
```

```python
df["gene_parsed"] = df["gene"].map(lambda x: "|".join([g.split("=")[-1] for g in x.split("|")[1:]]))
df["ortholog_parsed"] = df["ortholog"].map(lambda x: "|".join([g.split("=")[-1] for g in x.split("|")[1:]]))
df["value"] = 1.000
df.head(20)
```

```python
filename = "../data/orthology_related_files/pantherdb/PlantGenomeOrthologs_IRB_Modified.txt"
df[["gene_parsed","ortholog_parsed","value"]].to_csv(filename, index=False, header=False, sep="\t")
```
