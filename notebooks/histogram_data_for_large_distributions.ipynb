```python
import numpy as np
import json
with open("/Users/irbraun/Desktop/test.json", "w") as f:
    d = {"positives":list(np.random.normal(loc=0.2, size=100*100)), 
         "negatives":list(np.random.normal(loc=0.5, size=1000*1000))}
    json.dump(d, f)
```

```python
num_nodes_1 = 100
num_nodes_2 = 1000
pairwise_dist_1 = list(np.random.normal(loc=0.2, scale=0.1, size=num_nodes_1**2))
pairwise_dist_2 = list(np.random.normal(loc=0.5, scale=0.1, size=num_nodes_2**2))
```

```python
a = np.histogram(pairwise_dist_1, bins=100, range=(0,1), density=True)
a
b = np.histogram(pairwise_dist_2, bins=100, range=(0,1), density=True)
b
```

```python
import pandas as pd
num_bins = a[0].shape[0]
bin_width = 1/num_bins
bin_centers = [x+(bin_width/2) for x in a[1][:num_bins]]

df = pd.DataFrame({"bin_center":bin_centers, "frequency":a[0]})
df.head(20)
df.to_csv("/Users/irbraun/Desktop/test.csv", index=False)
df["approach"] = "aa"
df.head(20)
```

```python
import pandas as pd
num_bins = b[0].shape[0]
bin_width = 1/num_bins
bin_centers = [x+(bin_width/2) for x in b[1][:num_bins]]

df2 = pd.DataFrame({"bin_center":bin_centers, "frequency":b[0]})
df2.head(20)
df2.to_csv("/Users/irbraun/Desktop/test.csv", index=False)
df2["approach"] = "ab"
df2
```

```python
c = pd.concat([df,df2])
c.to_csv("/Users/irbraun/Desktop/test.csv", index=False)
c
```
