```python
import pandas as pd
```

```python
clinvar_path = "../../snpedia-nlp/genes_and_diseases.tsv"
snippets_path = "../../snpedia-nlp/genes_snps_snippets.tsv"
contexts_path = "../../snpedia-nlp/genes_snps_contexts.tsv"
```

```python
df = pd.read_csv(clinvar_path, sep="\t", header=0, names=["unique_gene_identifiers", "descriptions"])
df["species"] = "hsa"
df["other_gene_identifiers"] = ""
df["gene_models"] = ""
df["sources"] = "ClinVar"
df["annotations"] = ""
df.to_csv("../data/clinvar/clinvar_diseases.csv")
df
```

```python
df = pd.read_csv(snippets_path, sep="\t", header=0, names=["unique_gene_identifiers", "other_gene_identifiers", "descriptions"])
df["species"] = "hsa"
df["gene_models"] = ""
df["sources"] = "SNPedia"
df["annotations"] = ""
df.to_csv("../data/snpedia/snpedia_snippets.csv")
df
```

```python
df = pd.read_csv(contexts_path, sep="\t", header=0, names=["unique_gene_identifiers", "other_gene_identifiers", "descriptions"])
df["species"] = "hsa"
df["gene_models"] = ""
df["sources"] = "SNPedia"
df["annotations"] = ""
df.to_csv("../data/snpedia/snpedia_contexts.csv")
df
```
